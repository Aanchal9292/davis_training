```python
#to find out hcf
# input two number
num1 = int(input("Enter first number : "))
num2 = int(input("Enter second number : "))
#-----------------------
if(num2 == 0):
  print("HCF : ",num1)
else:
  while(num2 != 0):
    rem = num1%num2
    num1 = num2
    num2 = rem
  print("HCF : ",num1)
```

```python
num = int(input("Enter a number: "))

largest = 0

while num > 0:
    digit = num % 10      # get last digit
    if digit > largest:
        largest = digit
    num = num // 10       # remove last digit

print("Greatest digit is:", largest)
```

```python
# to find out factorial of a number
num = int(input("Enter a number :"))
if(num<0):
  print("Not possible")
elif(num==0):
  print("Factorial is 1")
else:
  fact = 1
  for x in range(1,num+1):
    fact = fact*x
  print("Factorial : ",fact)
```

```python
# to find out even factors of a number
num = int(input("Enter the number : "))
print("Even factors of", num, "are:")

for i in range(1, num + 1):
    if num % i == 0 and i % 2 == 0:
        print(i)
```
